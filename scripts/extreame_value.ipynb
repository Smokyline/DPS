```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def foo(x):
    return np.sin(x)/x

def K(delta_f, delta_x):
    return delta_f/delta_x

def deriv_foo(foo, x0, dx):
    df = foo(x0+dx) - foo(x0)
    return df/dx
```

```python
X = np.arange(0, 100., 0.5)
plt.plot(foo(X))
```

```python
delta = 0.5
df = [deriv_foo(foo, x, delta) for x in X]

plt.plot(foo(X))
plt.plot(df)
plt.grid(True)
#plt.savefig('test.png', bbox_inches='tight', dpi=1000)
```
